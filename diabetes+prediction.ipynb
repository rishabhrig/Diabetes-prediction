```python
# Imputing the NaN values for columns in accordance to their distribution:

# here we are filling NaN values in respective coloumns of  's_copy', to prevent the distortion of original dataset 's', as it can be used for future reference.

# [NOTE]:-Just like pandas dropna() method manage and remove Null values from a data frame, fillna() manages and let the user replace NaN values with some value of their own.

s_copy["Glucose"].fillna(s_copy["Glucose"].mean(),inplace=True)               #The fillna() method replaces the NULL values with a specified value.
                                                                              #The fillna() method returns a new DataFrame object unless the inplace parameter is set to True ,in that case the fillna() method does the replacing in the original DataFrame instead.              
    
# s_copy = s_copy["Glucose"].fillna(s_copy["Glucose"].mean())------method 2 without using inplace

s_copy["BloodPressure"].fillna(s_copy["BloodPressure"].mean(),inplace=True)

s_copy["SkinThickness"].fillna(s_copy["SkinThickness"].median(),inplace=True) # since the distribution is skewed and consists of outliers,we are replacing with median
s_copy["Insulin"].fillna(s_copy["Insulin"].median(),inplace=True)

s_copy["BMI"].fillna(s_copy["BMI"].median(),inplace=True)
```

```python
#Splitting the dataset into testing and training:---->

   # train_test_split() ----->

   # train_test_split is a function in Sklearn model selection for splitting data arrays into two subsets: for training data and for testing data.
   # With this function, you don't need to divide the dataset manually. By default, Sklearn train_test_split will make random partitions for the two subsets.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/3,random_state=42,stratify=y)

   # X,y :---> these are inputs such as lists, arrays, dataframes or matrices.
 
   # random_state:---> this parameter is used to control the shuffling applied to the data before applying the split, it acts like a seed. random_state simply sets a seed to the random generator, so that your train-test splits are always deterministic. If you don't set a seed, it is different each time.
 
   # train_size :---> this is a float value whose value ranges between 0.0 and 1.0. it represents the proportion of our train size. it’s default value is none.
 
   # stratify :---> to split the dataset into train and test sets in a way that preserves the same proportions of examples in each class as observed in the original dataset. This is called a stratified train-test split.
           #'stratify = y' :---> this stratify the data in same proportion as in 'y' i.e in prortion to (268 & 500) or(268/768=0.348, & 500/768=0.651)
           # this stratifivation can be observed below.
```

```python
# METHOD 1 :- Creating a Classifier using linear regression algorithm:--->

from sklearn.linear_model import LinearRegression  # Linear Regression is a machine learning algorithm based on supervised learning. It performs a regression task. Regression models a target prediction value based on independent variables. It is mostly used for finding out the relationship between variables and forecasting.


lin_model = LinearRegression()
lin_model.fit(X_train, y_train)                   # '.fit()' fits the linear regression model & train the model using the training sets.
```

```python
print(lin_model.score(X_test, y_test)) # Every model in Scikit-learn has a score method after being trained on the data, usually X_train, y_train.
                                       # When call score on classifiers like linear regression the '.score()' method computes the accuracy score by default (accuracy is #correct_preds / #all_preds), the score method does not need the actual predictions.
                                       # So, when called 'lin_model.score(X_test, y_test)' it makes predictions using X_test under the hood and uses those predictions to calculate accuracy score.
        
```

```python
y_pred = lin_model.predict(X_test)    # We can predict quantities with the finalized regression model by calling the predict () function on the finalized model
```

```python
y_pred
```

```python
# METHOD 2 :- Creating a Classifier using KNN algorithm----->

# KNN algorithm :- The k-nearest neighbors (KNN) algorithm is a simple, supervised machine learning algorithm that can be used to solve both classification and regression problems.
#[DOUBT]-classification & regression problem ?
from sklearn.neighbors import KNeighborsClassifier

test_acc=[]      # creating 2 empty lists.              
train_acc=[]


for i in range(1,15):
    knn = KNeighborsClassifier(i)                  #setting up a knn classifier
    
    knn.fit(X_train,y_train)                       #fitting the model
                                                 
    train_acc.append(knn.score(X_train,y_train))   # computing the accuracy for both the training and the test data by using '.score()' method & inserting the score values in 'train_acc' & 'test_acc' by using '.append()' function. 
    test_acc.append(knn.score(X_test,y_test))
```

```python
# Visualisation of the Scores of the model:

plt.figure(figsize=(14,8))
plt.title('k-NN Varying number of neighbors')

plt.plot(train_acc, linestyle = 'dotted', linewidth = 3,     #The plot() function in pyplot module of matplotlib library is used to make a 2D hexagonal binning plot of points x,y. 
         marker='o', markerfacecolor='orange', markersize=12, label = 'train_acc')  # plt.plot() here uses data in train_acc(list) to operate on.     
plt.plot(test_acc, linestyle = 'dotted', linewidth = 3,                             # The default linestyle while plotting data is solid linestyle in matplotlib. We can change this 'linestyle' by using linestyle or 'ls' argument of plot() method. 
         marker='o', markerfacecolor='red', markersize=12, label = 'test_acc')      # Following are the linestyles available in matplotlib:-
plt.xlabel('x - axis')   # naming the x axis                                        # Using linestyle Argument:--> Solid,Dashed,Dotted,Dashdot,None.
                                                                                    # Using ls Argument:--> ‘-‘, ‘:’, ‘–‘, ‘-.’, ‘ ‘ 
plt.ylabel('y - axis') # naming the y axis                                          # Line Width : The width of a line is known as line width. One can change the line width of a graph in matplotlib using 'linewidth' feature.  'lw' can also be used as short form
#plt.plot(train_acc, linestyle = 'dotted', linewidth = 3, label = 'train_acc')      # 'markers' are used to mark the points where the continuity of the line or curve is changing.
                                                                                    # markers can be of following types:---> '.' Point marker, 'o' Circle marker, ',' Pixel marker, 'v' triangle_down marker, '^' triangle_up marker,'<' triangle_left marker, '>' triangle_right marker, '1' tri_down marker
#plt.plot(test_acc, linestyle = 'dotted', linewidth = 3, label = 'test_acc')        # '2' tri_up marker, '3' tri_left marker, '4' tri_right marker, 's' square marker, 'p' pentagon marker, '*' star marker, 'h' hexagon1 marker, 'H' hexagon2 marker, '+' Plus marker.
plt.legend() #show a legend on the plot                                             # 'x' X marker, 'D' Diamond marker, 'd' thin_diamond marker, '|' vline marker, '_' hline marker.
                                                                                    # 'markerfacecolor' is used to change the marker color in the plot.
plt.show()                                                                          # Color code abbreviations that can be used along with the line styles:-
                                                                                    # 'b' blue, 'g' green, 'r' red, 'c' cyan, 'm' magenta, 'y' yellow, 'k' black, 'w' white.
```

```python
knn = KNeighborsClassifier(6) # From the given plot we get to observe that we get a maximum testing accuracy when K=7,
                                          # Hence create a KNN classifier for K=7 to get the score.
knn.fit(X_train,y_train)                  # fitting the model to train.
knn.score(X_test,y_test)      
```

```python
# Model Performance Analysis:---->
# confusion matrix:- A much better way to evaluate the performance of a classifier is to look at the confusion matrix. The general idea is to count the number of times instances of class A are classified as class B.
# Each row in a confusion matrix represents an actual class, while each column represents a predicted class.

from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'],margins=True)   # crosstab:- This method is used to compute a simple cross-tabulation of two (or more) factors. By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.
 # index(y_test) : array-like, Series, or list of arrays/Series, Values to group by in the rows.
 # columns(y_pred) : array-like, Series, or list of arrays/Series, Values to group by in the columns.
 # rownames : sequence, default None, If passed, must match number of row arrays passed.
 # colnames : sequence, default None, If passed, must match number of column arrays passed.
 # margins : bool, default False, Add row/column margins (subtotals).
 # Row indicates the actual values of data and columns indicate the predicted data.
```

```python
# Classification Report:- As the name suggests, it is the report which explains everything about the classification. This is the summary of the quality of classification made by the constructed ML model

from sklearn.metrics import classification_report   # Precision Score:- it is the accuracy of the positive predictions.
                                                            # TP - True positive; FP - False positive
print(classification_report(y_test,y_pred))                 # precision = TP/(TP+FP)                                                         
                                                    # Recall Score:- fraction of the positives correctly identified
                                                            # FN - False Negative
                                                            # Recall = TP/(TP+FN)
                                                    # F1 Score:- takes both precision and recall into account; F1 Score = 2(precision*recall)/(precision+recall)
                                                 
                                                    # Support: It is the total entries of each class in the actual dataset. It is simply the sum of rows for every class-i                    
                #[DOUBT]:- WHY GETTING ONLY AVG/TOTAL. check wrt the solved one.             
```

```python
# ROC CURVE:---->
# ROC(receiver operating curve) tells us how good the model can distinguish between two things.
# The ROC curve is produced by calculating and plotting the true positive rate against the false positive rate for a single classifier at a variety of thresholds.

from sklearn.metrics import roc_curve
y_pred_proba = knn.predict_proba(X_test)[:,1]        # The predict method is used to predict the actual class while predict_proba method can be used to infer the class probabilities (i.e. the probability that a particular data point falls into the underlying classes).
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)   # sincetarget is (0,1), then the classifier would output a probability matrix of dimension (N,2). The first index refers to the probability that the data belong to class 0, and the second refers to the probability that the data belong to class 1.These two would sum to 1.
                                                         # then output the result can be obtained by:--->probability_class_1 = model.predict_proba(X)[:, 1]
                                                         # If we have k classes, the output would be (N,k), just specify the probability of which class you want.


#y_pred_proba
#fpr, tpr, thresholds
```
